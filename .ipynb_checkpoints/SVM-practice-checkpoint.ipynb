# Single feature SVM

```python
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()

X = cancer_data.data
Y = cancer_data.target

# Always check for imbalances of data before starting

print(f"Input data size: {X.shape}")
print(f"Output data size: {Y.shape}")
print("Label names: ", cancer_data.target_names)
```

```python
n_pos = (Y == 1).sum()
n_neg = (Y == 0).sum()
print(f"{n_pos} positive samples and {n_neg} negative samples")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear', C=1, random_state=42)
clf.fit(X_train, Y_train)
```

```python
accuracy = clf.score(X_test, Y_test)
print(f"The accuracy is {accuracy*100:.1f}%")
```

# Multiple features SVM

```python
from sklearn.datasets import load_wine

wine_data = load_wine()
X = wine_data.data
Y = wine_data.target
print("Input size is ", X.shape)
print("Output size is ", Y.shape)

print("Label names: ", wine_data.target_names)
```

```python
n_class0 = (Y==0).sum()
n_class1 = (Y==1).sum()
n_class2 = (Y==2).sum()
print(f"{n_class0} class 0 samples, {n_class1} class 1 samples, {n_class2} class 2 samples")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42)

clf = SVC(kernel="linear", C=1.0, random_state=42)
clf.fit(X_train, Y_train)
```

```python
accuracy = clf.score(X_test, Y_test)
print(f"The accuracy is: {accuracy*100:.1f}%")
```

```python
from sklearn.metrics import classification_report

pred = clf.predict(X_test)
print(classification_report(Y_test, pred))
```

# Radial Basis Function (RBF) kernel / Gaussian kernel

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.c_[# negative class
     (.3, -.8),
     (-1.5, -1),
     (-1.3, -.8),
     (-1.1, -1.3),
     (-1.2, -.3),
     (-1.3, -.5),
     (-.6, 1.1),
     (-1.4, 2.2),
     (1, 1),
     # positive class
     (1.3, .8),
     (1.2, .5),
     (.2, -2),
     (.5, -2.4),
     (.2, -2.3),
     (0, -2.7),
     (1.3, 2.1)].T
Y = [-1] * 8 + [1] * 8
```

```python
gamma_option = [1, 2, 4]

for i, gamma in enumerate(gamma_option, 1):
    svm = SVC(kernel='rbf', gamma=gamma)
    svm.fit(X, Y)
    plt.scatter(X[:, 0], X[:, 1], c=['b']*8+['r']*8, zorder=10, cmap=plt.cm.Paired)
    plt.axis("tight")
    XX, YY = np.mgrid[-3:3:200j, -3:3:200j]
    Z = svm.decision_function(np.c_[XX.ravel(), YY.ravel()])
    Z = Z.reshape(XX.shape)
    plt.pcolormesh(XX, YY, Z>0, cmap=plt.cm.Paired)
    plt.contour(XX,YY,Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'], levels=[-.5, 0, .5])
    plt.title('gamma = %d'%gamma)
    plt.show()
```
